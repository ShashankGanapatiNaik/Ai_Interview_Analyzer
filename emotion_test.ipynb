```python
import cv2
#start webcam
cap=cv2.VideoCapture(0)
#0 → First camera (usually laptop webcam)
#1 → External USB camera
#2 → Second external camera
print("Press q for exit")
while True:
    ret,frame=cap.read()
    #Boolean value (True or False)
    #True → Frame captured successfully
    #False → Failed to capture frame
    if not ret:
        break
    try:
        #analyze emotion
        result=DeepFace.analyze(frame,actions=["emotion"],enforce_detection=False)
        #Checks if face is clearly detected. If no face → throws error
        emotion=result[0]["dominant_emotion"]

        #put emotion text on frame (live video)
        cv2.putText(frame,emotion,(50,50),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1,(0,255,0),2)
        #FONT_HERSHEY_SIMPLEX-->Font style. 1->font size,2->thicknexs
    except:
        pass
    cv2.imshow("Live emotion detection",frame)
    #This shows the video window.
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
    #It means: “Wait 1 millisecond for a key press.” If user presses a key: It returns ASCII value of that key.
cap.release()
cv2.destroyAllWindows()
```

```python
from collections import Counter
video_path="interview.mp4"
cap=cv2.VideoCapture(video_path)
fps=cap.get(cv2.CAP_PROP_FPS)
frame_interval=int(fps)
frame_count=0
emotion_list=[]
print("Processing video")
while True:
    ret,frame=cap.read()
    if not ret:
        break
    if frame_count%frame_interval==0:
        try:
            result=DeepFace.analyze(frame,actions=["emotion"],enforce_detection=False)
            emotion=result[0]["dominant_emotion"]
            emotion_list.append(emotion)
            print(f"Detected:{emotion}")
        except:
            pass
    frame_count+=1
cap.release()
print("\nVideo processing completed")

#calulate result
emotion_counts=Counter(emotion_list)
total=sum(emotion_counts.values())
print("\nEmotion Distribution")
for emotion,count in emotion_counts.items():
    percentage=(count/total)*100
    print(f"{emotion}:{percentage:.2f}%")
if emotion_counts:
    dominant=emotion_counts.most_common(1)[0][0]
    print(f"\nDominant Emotion: {dominant}")
else:
    print("No emotions detected.")
```
